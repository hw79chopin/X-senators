```python
import pandas as pd
from glob import glob
```

```python
dict_data = {}
for i in glob('data/*.csv'):
    try:
        dict_data[i.split('/')[1].split('.csv')[0]] = pd.read_csv(i)
    except:
        dict_data[i.split('/')[1].split('.csv')[0]] = pd.read_csv(i, encoding='cp949')
```
